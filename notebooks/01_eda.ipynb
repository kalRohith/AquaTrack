```python
## STEP 1: INLINE EDA SETUP (NO EXTERNAL MODELS IMPORTS)
# This cell is intentionally verbose with comments so each transformation is explicit.

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Optional seaborn styling: fallback to matplotlib if seaborn is unavailable.
try:
    import seaborn as sns
    sns.set_theme(style="whitegrid", context="talk")
    HAS_SEABORN = True
except Exception:
    HAS_SEABORN = False

# ---------- Constants ----------
PROJECT_ROOT = Path("..")
DATA_PATH = PROJECT_ROOT / "aquatrack_labeled.csv"
ARTIFACT_DIR = PROJECT_ROOT / "models" / "artifacts"
PLOT_DIR = ARTIFACT_DIR / "plots"

TARGET_COL = "dehydration_risk"
OLD_TARGET_COL = "dehydration_level"
SKEW_THRESHOLD = 1.5
HIGH_CORR_THRESHOLD = 0.85
PAIRPLOT_TOP_K = 8

# Create output folders once so all plot saves succeed.
ARTIFACT_DIR.mkdir(parents=True, exist_ok=True)
PLOT_DIR.mkdir(parents=True, exist_ok=True)

# ---------- Data loading ----------
df = pd.read_csv(DATA_PATH)

# Align old/new target naming so downstream cells use one canonical name.
if OLD_TARGET_COL in df.columns and TARGET_COL not in df.columns:
    df = df.rename(columns={OLD_TARGET_COL: TARGET_COL})

if TARGET_COL not in df.columns:
    raise KeyError(f"Target column '{TARGET_COL}' not found in dataset.")

# ---------- Helper functions ----------
def high_correlation_pairs(corr: pd.DataFrame, thresh: float = 0.85) -> pd.DataFrame:
    """Return feature pairs whose absolute correlation exceeds threshold."""
    pairs = []
    for i, left in enumerate(corr.columns):
        for right in corr.columns[i + 1:]:
            r = corr.loc[left, right]
            if np.isfinite(r) and abs(r) >= thresh:
                pairs.append((left, right, float(r)))
    out = pd.DataFrame(pairs, columns=["feature_a", "feature_b", "r"])
    if out.empty:
        return out
    return out.sort_values("r", key=np.abs, ascending=False).reset_index(drop=True)


def iqr_outlier_counts(frame: pd.DataFrame, cols: list[str], factor: float = 1.5) -> pd.Series:
    """Count outliers per column using IQR fences."""
    counts = {}
    for col in cols:
        s = pd.to_numeric(frame[col], errors="coerce").dropna()
        if len(s) < 4:
            counts[col] = 0
            continue
        q1, q3 = s.quantile(0.25), s.quantile(0.75)
        iqr = q3 - q1
        lo, hi = q1 - factor * iqr, q3 + factor * iqr
        counts[col] = int(((s < lo) | (s > hi)).sum())
    return pd.Series(counts).sort_values(ascending=False)


def skewness_report(frame: pd.DataFrame, cols: list[str]) -> pd.Series:
    """Compute skewness and rank by absolute magnitude."""
    return frame[cols].apply(pd.to_numeric, errors="coerce").skew().sort_values(key=np.abs, ascending=False)


def _save_show(fig, out_path: Path):
    """Save and show figure with consistent export quality."""
    fig.tight_layout()
    fig.savefig(out_path, dpi=140, bbox_inches="tight")
    plt.show()


# ---------- EDA: overview ----------
print("Shape:", df.shape)
print("\nInfo:")
df.info()
print("\nDescribe:")
display(df.describe(include="all").T)

display(df.head(5))
display(df.tail(3))

# ---------- Missingness ----------
missing = df.isnull().sum()
missing_nonzero = missing[missing > 0].sort_values(ascending=False)
if not missing_nonzero.empty:
    fig, ax = plt.subplots(figsize=(12, 5))
    if HAS_SEABORN:
        sns.barplot(x=missing_nonzero.index, y=missing_nonzero.values, ax=ax, color="#8e44ad")
    else:
        ax.bar(missing_nonzero.index.astype(str), missing_nonzero.values, color="#8e44ad")
    ax.set_title("Missing Values Per Column")
    ax.set_xlabel("Columns")
    ax.set_ylabel("Null Count")
    ax.tick_params(axis="x", rotation=75)
    _save_show(fig, PLOT_DIR / "eda_main_missing_values.png")

# ---------- Distribution analysis ----------
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
skews = skewness_report(df, numeric_cols)
print("Skewed columns (|skew| > 1.5):")
print(skews[skews.abs() > SKEW_THRESHOLD])

ncols = 3
nrows = int(np.ceil(len(numeric_cols) / ncols))
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(7 * ncols, 4 * nrows))
axes = np.array(axes).reshape(-1)

for i, col in enumerate(numeric_cols):
    ax = axes[i]
    s = pd.to_numeric(df[col], errors="coerce").dropna()
    if HAS_SEABORN:
        sns.histplot(s, kde=True, ax=ax, color="#3498db")
    else:
        ax.hist(s, bins=30, color="#3498db", alpha=0.85)
    skew_val = float(s.skew()) if len(s) > 2 else 0.0
    ax.set_title(f"Distribution: {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")
    ax.text(0.02, 0.95, f"skew={skew_val:.2f}", transform=ax.transAxes, va="top")

for j in range(i + 1, len(axes)):
    axes[j].axis("off")

_save_show(fig, PLOT_DIR / "eda_main_distributions.png")

# ---------- Correlation analysis ----------
bio_cols = [c for c in numeric_cols if ("sweat" in c.lower() or "salivary" in c.lower())]
if len(bio_cols) >= 2:
    corr = df[bio_cols].corr(numeric_only=True)
    mask = np.triu(np.ones_like(corr, dtype=bool))

    fig, ax = plt.subplots(figsize=(13, 10))
    if HAS_SEABORN:
        sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", mask=mask, ax=ax)
    else:
        im = ax.imshow(corr.values, cmap="coolwarm")
        fig.colorbar(im, ax=ax)
        ax.set_xticks(range(len(corr.columns)))
        ax.set_xticklabels(corr.columns, rotation=90)
        ax.set_yticks(range(len(corr.index)))
        ax.set_yticklabels(corr.index)
    ax.set_title("Sweat + Saliva Correlation Heatmap")
    ax.set_xlabel("Features")
    ax.set_ylabel("Features")
    _save_show(fig, PLOT_DIR / "eda_main_corr_sweat_saliva.png")

    display(high_correlation_pairs(corr, thresh=HIGH_CORR_THRESHOLD))

# ---------- Pairplot (top variance features) ----------
if HAS_SEABORN:
    pair_cols = [c for c in numeric_cols if c != TARGET_COL]
    top_var_cols = df[pair_cols].var().sort_values(ascending=False).head(PAIRPLOT_TOP_K).index.tolist()
    pair_df = df[top_var_cols + [TARGET_COL]].dropna().copy()
    pp = sns.pairplot(pair_df, hue=TARGET_COL, diag_kind="kde")
    pp.fig.suptitle("Pairplot by dehydration_risk (Top variance features)", y=1.02)
    pp.savefig(PLOT_DIR / "eda_main_pairplot.png", dpi=140)
    plt.show()

# ---------- Target analysis ----------
fig, ax = plt.subplots(figsize=(8, 4))
if HAS_SEABORN:
    sns.countplot(data=df, x=TARGET_COL, palette="viridis", ax=ax)
else:
    counts = df[TARGET_COL].astype(str).value_counts()
    ax.bar(counts.index, counts.values, color="#2ecc71")
ax.set_title("dehydration_risk Class Distribution")
ax.set_xlabel(TARGET_COL)
ax.set_ylabel("Count")
_save_show(fig, PLOT_DIR / "eda_main_target_countplot.png")

# Show class-level feature means to support interpretability.
mean_table = df.groupby(TARGET_COL)[[c for c in numeric_cols if c != TARGET_COL]].mean(numeric_only=True)
display(mean_table)

# Rank features by ANOVA p-value against target classes.
anova_rows = []
for col in [c for c in numeric_cols if c != TARGET_COL]:
    groups = [grp[col].dropna().values for _, grp in df.groupby(TARGET_COL)]
    if all(len(g) > 1 for g in groups):
        _, p = f_oneway(*groups)
        anova_rows.append((col, float(p)))

anova_df = pd.DataFrame(anova_rows, columns=["feature", "anova_p_value"]).sort_values("anova_p_value")
display(anova_df)

# Optional outlier summary table for quick QA.
outlier_table = iqr_outlier_counts(df, [c for c in numeric_cols if c != TARGET_COL], factor=1.5)
display(outlier_table.to_frame("iqr_outlier_count"))

print("EDA complete. All generated figures saved to:", PLOT_DIR)
```

## Key observations (brief)
- **Target pipeline (notebook 03):** six **primary** biomarkers → median + **RobustScaler** + KMeans → categorical **`dehydration_level`** (domain ordering via z-score sum on centroids, not fixed thresholds).
- **Missingness** — global **median** is standard for skewed labs; subject-wise `ffill` is an alternative for longitudinal rows.
- **Heavy-tailed markers** — prefer **RobustScaler** / IQR caps; **StandardScaler** after log-transform is another standard option (see notebook 02).
- **Repeated measures** per `id` — **grouped** train/validation splits.
- **Class imbalance** in `dehydration_level` is possible with small *n*; see notebook 03.
